```python

df_copy=pd.read_csv("https://d2beiqkhq929f0.cloudfront.net/public_assets/assets/000/001/125/original/aerofit_treadmill.csv?1639992749")
df_copy['Product'].replace(['KP281', 'KP481', 'KP781'], [0, 1, 2], inplace=True)
df_copy['Gender'].replace(['Female', 'Male'], [1, 0], inplace=True)
df_copy['MaritalStatus'].replace(['Single', 'Partnered'], [0, 1], inplace=True)
df_copy.corr()
sns.heatmap(df_copy.corr(),annot=True)
```

```python
age_bins=[17,21,31,41,51]
labels_age=["20","20-30","30-40","40-50"]
df["age_bin"]=pd.cut(df["Age"],bins=age_bins,labels=labels_age)

income_bins=[25000,50000,75000,100000,125000]
labels_income=["25k-50k","50k-75k","75k-1l","1l-1.25l"]
df["Income_bins"]=pd.cut(df["Income"],bins=income_bins,labels=labels_income)

miles_bin=[0,50,100,150,200,250,300,350,400]
df["Miles_bin"]=pd.cut(df["Miles"],bins=miles_bin,)
df.head()
```
