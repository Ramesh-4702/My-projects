```python
df['date_added'] = pd.to_datetime(df['date_added'])
df['month_added'] = df['date_added'].dt.month.fillna(0)
month_df = df[df['month_added']>0]
month_df = month_df[month_df['release_year']>0]
month_df = month_df[['type',"release_year",'month_added']]
tv_Shows_per_month = month_df[month_df['type']=='TV Show']
new_tv_df = tv_Shows_per_month.groupby(['release_year','month_added'])['type'].count()
new_tv_df = pd.DataFrame(new_tv_df)
new_tv_df.reset_index(inplace = True)
new_tv_df = new_tv_df.pivot('month_added','release_year','type').fillna(0)
sns.heatmap(new_tv_df)
```

```python
movies_per_month = month_df[month_df['type']=='Movie']
new_tv_df = movies_per_month.groupby(['release_year','month_added'])['type'].count()
new_tv_df = pd.DataFrame(new_tv_df)
new_tv_df.reset_index(inplace = True)
new_tv_df = new_tv_df.pivot('month_added','release_year','type').fillna(0)
sns.heatmap(new_tv_df)
```

```python


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("https://d2beiqkhq929f0.cloudfront.net/public_assets/assets/000/000/940/original/netflix.csv")

from wordcloud import WordCloud, ImageColorGenerator
text = " ".join(str(each) for each in df.director)
# Create and generate a word cloud image:
wordcloud = WordCloud(max_words=200, background_color="gray").generate(text)
plt.figure(figsize=(10,6))
plt.figure(figsize=(15,10))
# Display the generated image:
plt.imshow(wordcloud, interpolation='Bilinear')
plt.title('Most Popular Directors',fontsize = 30)
plt.axis("off")
plt.show()
```
